# Skip-gram

Skip-gram is the second main architecture used by Word2Vec. Its goal is the exact opposite of CBOW: it tries to **predict the surrounding context words given a single target word.**


### How it Works: The Working Principle

Instead of averaging context words to find the middle word, Skip-gram takes the middle word and uses it to predict all the words that appear in its neighborhood (based on the window size).

#### The Architecture
1. **Input Layer:** The model takes the one-hot encoded vector of the single target word.
2. **Hidden Layer:** The input is multiplied by a weight matrix. There is **no averaging** in Skip-gram because there is only one input word.
3. **Output Layer:** The network outputs multiple predictions (one for each context position) using the Softmax function.


### Detailed Working with an Example

Let's use the same sentence we used for CBOW:
**"The movie was very good"**

*   **Target Word:** "very"
*   **Context Words (Window Size = 1):** ["was", "good"]

The model will split this into multiple independent training pairs: 
*   (Input: "very", Output: "was")
*   (Input: "very", Output: "good")

It passes "very" through the network and updates the weights until it gets good at predicting both "was" and "good".

```mermaid
graph LR
    subgraph Input Layer
        IN["Target Word<br>'very'<br>[0, 0, 0, 1, 0]"]
    end

    subgraph Hidden Layer
        WM["Hidden Nodes<br>(Weight Matrix)"]
    end

    subgraph Output Layer (Predictions)
        OUT1["Context Word 1<br>'was'"]
        OUT2["Context Word 2<br>'good'"]
    end

    IN --> WM
    WM --> OUT1
    WM --> OUT2



### Pros and Cons of Skip-gram

#### Pros (Advantages)
1. **Excellent with Rare Words:** Because Skip-gram treats every (target, context) pair as a separate training example, rare words are not "averaged out" or lost. It learns highly accurate representations for infrequent terms.
2. **Captures Finer Semantics:** It preserves specific one-to-one word relationships much better than CBOW.

#### Cons (Disadvantages)
1. **Slower to Train:** Since it processes multiple context predictions for every single target word, it takes much longer to train than CBOW.
2. **Higher Resource Usage:** It uses more memory and computational power, especially on very large datasets.


### Implementation Code (Gensim)

In Gensim, you select the Skip-gram architecture by setting the `sg` parameter to `1`.

```python
from gensim.models import Word2Vec

sentences = [
    ["the", "movie", "was", "very", "good"],
    ["nlp", "is", "a", "fascinating", "field"]
]

# sg=1 tells Gensim to use Skip-gram instead of CBOW
model = Word2Vec(
    sentences, 
    vector_size=100, 
    window=5, 
    min_count=1, 
    sg=1, 
    workers=4
)


---
---

### When to use CBOW or Skip-gram

Choosing between CBOW and Skip-gram depends primarily on the size of your dataset, your computational resources, and whether your focus is on frequent or rare words.
#### Use CBOW When:
1. **Speed is vital:** You need to train the model quickly. Because CBOW averages the context, it requires far fewer mathematical operations and trains significantly faster.
2. **You have a massive dataset:** With huge amounts of data, the averaging effect of CBOW smooths over noise and produces very stable vectors.
3. **Frequent words are the priority:** You mostly care about the most common words in your corpus having strong, generalized representations.
4. **General meaning over specific pairs:** You want the model to capture the general "theme" of contexts rather than highly specific word-to-word relationships.

#### Use Skip-gram When:
1. **You have a smaller dataset:** Skip-gram extracts more "training pairs" out of every sentence (because one target word predicts multiple contexts). It performs much better than CBOW when data is limited.
2. **Rare words are important:** Because Skip-gram doesn't average anything out, a rare word gets a direct update against its context words. It produces vastly superior vectors for words that appear infrequently.
3. **Finer semantic details matter:** You need the model to understand specific analogies and nuanced, one-to-one word relationships.
4. **Time and Compute are not an issue:** You are okay with training taking significantly longer and using more memory.


### The Quick Rule of Thumb
*   **CBOW:** Large Dataset + Need Speed + Frequent Words.
*   **Skip-gram:** Small Dataset + Need Precision + Rare Words.
