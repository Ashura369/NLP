# word2vec

Word2Vec is a popular technique in NLP used to create **Word Embeddings**. Unlike Bag of Words or TF-IDF, which create sparse, high-dimensional vectors, Word2Vec creates **dense, low-dimensional vectors** (typically 100 to 300 dimensions) where words with similar meanings are placed close together in a mathematical space.

### The Core Idea
Word2Vec is based on the **Distributional Hypothesis**, which states that words appearing in similar contexts tend to have similar meanings. For example, if "pizza" and "burger" both appear frequently near the word "delicious" or "eat," the model will learn that they are related.


### Two Main Architectures

Word2Vec uses a simple neural network with one hidden layer to learn these vectors. There are two ways to train it:

#### 1. CBOW (Continuous Bag of Words)
*   **Goal:** Predict a target word based on its surrounding context words.
*   **Example:** Given ["The", "cat", "___", "on", "the", "mat"], the model tries to predict "sat."
*   **Best for:** Faster training and works well with frequent words.

#### 2. Skip-gram
*   **Goal:** Predict the surrounding context words based on a single target word.
*   **Example:** Given "sat," the model tries to predict ["The", "cat", "on", "the", "mat"].
*   **Best for:** Working with small datasets and representing rare words more accurately.


### Key Feature: Semantic Relationships
Because Word2Vec represents words as points in a multi-dimensional space, it can perform "vector arithmetic" to capture complex relationships.
*   **Classic Example:** Vector("King") - Vector("Man") + Vector("Woman") ≈ Vector("Queen")


### Pros and Cons of Word2Vec

#### Pros (Advantages)
1.  **Captures Semantic Meaning:** It understands that "Happy" and "Joyful" are similar, whereas TF-IDF treats them as completely different words.
2.  **Dense Vectors:** Instead of a 50,000-dimensional vector of zeros, it uses a 300-dimensional vector of real numbers. This is much more memory-efficient.
3.  **Reduces Dimensionality:** It compresses the vast vocabulary into a small, manageable space without losing information.
4.  **Generalization:** Because it understands similarity, a model trained on Word2Vec can handle synonyms it hasn't seen frequently in the specific training labels.

#### Cons (Disadvantages)
1.  **Context-Independent:** This is the biggest drawback. Each word has only **one** vector. The word "Bank" has the same vector whether you are talking about a "river bank" or a "savings bank."
2.  **Large Training Data Needed:** To learn good relationships, Word2Vec requires a massive amount of text (millions of sentences).
3.  **Out of Vocabulary (OOV):** Like previous methods, if it encounters a word it wasn't trained on, it cannot generate a vector for it.
4.  **Complexity:** It is more computationally expensive to train than simple statistical methods like TF-IDF.


Word2Vec was a revolutionary step in NLP because it allowed machines to understand the "meaning" and "relationships" between words for the first time. However, it has been largely superseded by context-aware models like **BERT**, which can generate different vectors for the same word depending on the sentence.

```python
import pandas
import numpy

import gensim

import warnings
warnings.filterwarnings('ignore')
```

```python
import gensim.downloader as api
```

```python
from gensim.models import Word2Vec, KeyedVectors
```

```python
# downloading "Google news Word2Vec model" directly from gensim, 1.6 gb

    # It is a "Pre-trained" model. This means you don't have to spend days training it yourself. Google already did the hard work using their massive supercomputers.

    # Trained on: About 100 billion words from Google News articles.
    # Vocabulary: It knows 3 million unique words and phrases.
    # Dimensions: Each word is converted into a vector of 300 numbers.
    # Architecture: It was trained using the Skip-gram method.

model = api.load('word2vec-google-news-300')
```

#### Using some common methods to in the google news dataset

```python
# printing most similar words

print(list(model.most_similar('lion')))
print(list(model.most_similar('king')))
```

```python
# checking how close two words are

print(model.similarity('king', 'queen'))
print(model.similarity('king', 'lion'))
print(model.similarity('rat', 'cat'))
print(model.similarity('male', 'strength'))
print(model.similarity('female', 'weak'))
```

```python
# finding the odd one out in the given list

print(model.doesnt_match(["breakfast", "cereal", "dinner", "lunch", "car"]))
print(model.doesnt_match(['sing','song','singer','writer']))
```

```python
# prints distance btwn two words
# used mostly in Kmeans, KNN

print(model.distance('media', 'news'))
print(model.distance('king', 'queen'))
```

```python
# Returns the actual 300-dimensional numerical vector (the list of numbers) for that word.

print(model.get_vector('office'))
```

#### Using `distance` in KNN

```python
import numpy as np
from sklearn.neighbors import NearestNeighbors
import gensim.downloader as api

# 1. Load the pre-trained Word2Vec model
wv = api.load('word2vec-google-news-300')

# 2. Let's pick a list of words to "search" through
words_to_index = ["apple", "orange", "car", "truck", "bike", "banana", "bus"]
word_vectors = [wv[word] for word in words_to_index]

# 3. Initialize the KNN model
# We use 'cosine' metric because word vectors work best with cosine distance
knn = NearestNeighbors(n_neighbors=3, metric='cosine')

# 4. Fit the model with our word vectors
knn.fit(word_vectors)

# 5. Let's find the 3 nearest neighbors for the word "bicycle"
# First, get the vector for "bicycle"
target_vector = wv['bicycle'].reshape(1, -1)

# Find the distances and indices of the neighbors
distances, indices = knn.kneighbors(target_vector)

# 6. Print the results
print("Results for 'bicycle':")
for i in range(len(distances[0])):
    word = words_to_index[indices[0][i]]
    dist = distances[0][i]
    print(f"Word: {word}, Distance: {dist:.4f}")
```
