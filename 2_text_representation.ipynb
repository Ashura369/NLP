# One Hot Encoding (OHE)

One Hot Encoding is a process used to convert **Categorical Data** (words/labels) into a **Numerical Format** that Machine Learning models can understand.

### Why do we need it?
Machines can't understand text. If we use numbers like `1, 2, 3` for cities, the model might think **City 3 > City 1**. One Hot Encoding treats every category as equal.



### How it Works
It creates a new binary column (0 or 1) for every unique category in your data.

**Original Data:**
| User ID | City |
| :--- | :--- |
| 1 | Delhi |
| 2 | Mumbai |
| 3 | Pune |

**After One Hot Encoding:**
| User ID | City_Delhi | City_Mumbai | City_Pune |
| :--- | :--- | :--- | :--- |
| 1 | **1** | 0 | 0 |
| 2 | 0 | **1** | 0 |
| 3 | 0 | 0 | **1** |



### Python Code Example

```python
import pandas as pd

# 1. Create a dummy dataset
data = {'City': ['Delhi', 'Mumbai', 'Pune', 'Delhi']}
df = pd.DataFrame(data)

# 2. Apply One Hot Encoding using Pandas
ohe_df = pd.get_dummies(df, columns=['City'])

# 3. View the result
print(ohe_df)


## Pros and Cons of One Hot Encoding in NLP

In NLP, One Hot Encoding represents each word in your vocabulary as a unique vector where only one element is 1 and all others are 0. The size of the vector is equal to the total number of unique words in your corpus.

### Pros (Advantages)
1.  **Simplicity:** It is the most intuitive way to convert text into numbers. It is easy to understand and implement for small vocabularies.
2.  **No Mathematical Bias:** It does not assume any order between words. For example, it does not suggest that the word "King" (assigned index 10) is mathematically "greater" than "Queen" (assigned index 5).
3.  **Baseline Model:** It serves as a good starting point for simple text classification tasks using algorithms like Naive Bayes.

### Cons (Disadvantages)
1.  **Extreme Sparsity:** In NLP, vocabularies are often huge (e.g., 50,000 to 100,000 words). This means every single word is represented by a vector of 100,000 dimensions where 99,999 entries are zero. This is highly inefficient for memory and computation.
2.  **No Semantic Relationship:** This is the biggest drawback in NLP. In One Hot Encoding, the distance between any two words is exactly the same. The model cannot tell that "Run" is more related to "Jog" than it is to "Apple." All words are treated as completely independent entities.
3.  **Out of Vocabulary (OOV) Issues:** If the model encounters a new word during testing that wasn't in the training vocabulary, it cannot represent it at all.
4.  **Fixed Representation:** The representation of a word is static and does not change based on context. For example, the word "Bank" would have the same vector whether it refers to a river bank or a financial bank.


While One Hot Encoding was the standard in early NLP, it has largely been replaced by **Word Embeddings** (like Word2Vec, GloVe, or FastText) and **Transformers**, which represent words in a dense, lower-dimensional space that captures meaning and relationships.

## Bag of Words (BoW)

The Bag of Words model converts text into numerical representations by following these four steps:

#### 1. Text Normalization
Before building the model, the text is usually preprocessed. This includes converting all text to lowercase, removing punctuation, and sometimes removing stop words (like "is", "the", "at").

#### 2. Building the Vocabulary
The model looks at all the words present in all the documents (the corpus) and creates a list of all unique words. This list is your vocabulary.
*Example:* 
Sentence A: "I love NLP"
Sentence B: "NLP is great"
Vocabulary: {I, love, nlp, is, great}

#### 3. Vectorization
For each document, a vector is created with a length equal to the number of unique words in the vocabulary. Each position in the vector represents a specific word from the vocabulary.

#### 4. Counting Word Frequency
The model fills the vector by counting how many times each word from the vocabulary appears in that specific document.

### Step-by-Step Example
**Documents:**
- Document 1: "the cat sat"
- Document 2: "the dog sat"
**Step 1: Vocabulary**
{cat, dog, sat, the} (Size = 4)
**Step 2: Vector Representation**
Each vector will have 4 slots: [cat, dog, sat, the]
**Step 3: Final Vectors**
- Document 1: [1, 0, 1, 1] 
  (Explanation: 'cat' appears once, 'dog' appears zero times, 'sat' appears once, 'the' appears once)
- Document 2: [0, 1, 1, 1]
  (Explanation: 'cat' appears zero times, 'dog' appears once, 'sat' appears once, 'the' appears once)

### Pros and Cons of Bag of Words

#### Pros (Advantages)
1.  **Simplicity:** The concept is easy to understand and mathematically straightforward to implement.
2.  **Captures Frequency:** It provides a numerical value based on how often a word appears, which is very useful for basic text classification tasks.
3.  **No Pre-trained Models Needed:** Unlike Word Embeddings, you do not need a pre-trained model to use Bag of Words; it is built entirely from your own data.
4.  **Effective for Simple Tasks:** For tasks like spam detection or simple sentiment analysis, it often provides a strong baseline performance.

#### Cons (Disadvantages)
1.  **Loss of Sequence and Order:** The most significant drawback is that it ignores the order of words in a sentence. Hence it no longer becomes able to convey the same meaning as original text. This means the model loses the context and grammar of the sentence.  
2.  **Out of Vocabulary (OOV):** It can only handle words it has seen during the "fitting" process. It cannot represent new words (Out-of-Vocabulary words) during testing. **Hence as it sees a new word which is not available in its dictionary, it will completely ignore that word, leading to potential loss of important data.**
3.  **Data Sparsity:** In large vocabularies, the resulting vectors are mostly zeros, which is computationally expensive and memory-intensive.
4.  **Semantic Ignorance:** It cannot identify that two different words might have the same meaning (e.g., "fast" and "quick").
5.  **Frequent Word Bias:** Highly frequent but meaningless words (like "of", "and", "the") can dominate the vectors, potentially misleading the model.
6. For eg, if there are two sentences like **"It is a good movie"** and another one is **"It is not a good movie"**. According to BoW these two sentences convey the very close meaning as most of the words are being repeated. But in actually these two sentences convey very different meanings.

# CountVectorizer Hyperparameters

The `CountVectorizer` class in scikit-learn has several parameters that allow you to control how your text is converted into numerical vectors.

### Key Hyperparameters

| Parameter | Type | Default | Description |
| :--- | :--- | :--- | :--- |
| **lowercase** | boolean | True | Converts all characters to lowercase before tokenizing. |
| **stop_words** | string or list | None | If set to 'english', it removes common English words (the, a, is). You can also pass a custom list of words to ignore. |
| **ngram_range** | tuple | (1, 1) | The lower and upper boundary of the range of n-values for different n-grams. (1, 1) means unigrams, (1, 2) means unigrams and bigrams. |
| **max_features** | int | None | If not None, it only keeps the top 'N' most frequent words in the vocabulary. This helps control the size of your vector. |
| **max_df** | float or int | 1.0 | Ignore terms that have a document frequency higher than the threshold. 0.9 means ignore words that appear in more than 90% of documents. |
| **min_df** | float or int | 1 | Ignore terms that have a document frequency lower than the threshold. 2 means a word must appear in at least 2 documents to be kept. |
| **analyzer** | string | 'word' | Whether the feature should be made of word n-grams or character n-grams. Options are 'word', 'char', 'char_wb'. |
| **binary** | boolean | False | If True, all non-zero counts are set to 1. This is useful for models that only care if a word exists or not, rather than its frequency. |



### Example Code: Using Hyperparameters

This example demonstrates how to combine these parameters to create a more efficient and meaningful Bag of Words model.

```python
from sklearn.feature_extraction.text import CountVectorizer

# Example dataset
text_data = [
    "The movie was great and the acting was amazing.",
    "The movie was bad, the acting was terrible.",
    "I loved the movie, it was a great experience."
]

# Initializing CountVectorizer with custom hyperparameters
cv = CountVectorizer(
    lowercase=True,                # Convert to lowercase
    stop_words='english',          # Remove common English words
    ngram_range=(1, 2),            # Use both single words and pairs of words
    max_features=10,               # Keep only the top 10 most frequent features
    min_df=1,                      # Word must appear in at least 1 document
    max_df=0.8                     # Ignore words appearing in >80% of documents
)

# Fit and transform the data
X = cv.fit_transform(text_data)

# See the features (words/n-grams) created
print("Vocabulary:", cv.get_feature_names_out())

# See the vectors
print("Vectors:\n", X.toarray())

## N-Grams

Quite similary to BoW.


In Natural Language Processing, an N-gram is a contiguous sequence of **n** items from a given sample of text or speech. These items can be phonemes, syllables, letters, or, most commonly, words.

### Types of N-grams
*   **Unigram (n=1):** Individual words. Example: "The", "movie", "was", "good".
*   **Bigram (n=2):** Pairs of consecutive words. Example: "The movie", "movie was", "was good".
*   **Trigram (n=3):** Triplets of consecutive words. Example: "The movie was", "movie was good".



### How it Works: The Sliding Window
N-grams work by sliding a window of size **n** over a sentence, one word at a time.

**Example Sentence:** "The food is not good"

**If n=2 (Bigrams), the windows are:**
1.  [The, food]
2.  [food, is]
3.  [is, not]
4.  [not, good]

By grouping words together, the model can capture small amounts of local context that individual words (Unigrams) miss.

### Implementation in CountVectorizer
You can implement N-grams using the `ngram_range` parameter:

```python
from sklearn.feature_extraction.text import CountVectorizer

# ngram_range=(1, 2) means use both Unigrams and Bigrams
cv = CountVectorizer(ngram_range=(1, 2))
X = cv.fit_transform(["The food is not good"])

print(cv.get_feature_names_out())



### Pros and Cons of N-grams

#### Pros (Advantages)
1.  **Captures Context:** Unlike Bag of Words (which treats words individually), N-grams capture the relationship between adjacent words. Captures the semantic meaning of the sentence.
2.  **Handles Negation:** In sentiment analysis, a Unigram model might see "good" and think it is positive. A Bigram model sees "not good" and correctly identifies it as negative.
3.  **Captures Phrases:** It can identify common phrases or names like "New York" or "Artificial Intelligence" as a single unit rather than separate words.

#### Cons (Disadvantages)
1.  **Dimensionality Explosion:** This is the biggest drawback. If your vocabulary has 10,000 words, you have 10,000 Unigrams. However, you could potentially have 100 million Bigrams (10,000 squared). This makes the vector extremely large.
2.  **Sparsity:** Most possible N-gram combinations never appear in your text. This results in vectors with a massive number of zeros, which is inefficient for memory.
3.  **Fixed Context:** N-grams only capture context within the window size. If a word is related to another word 10 words away, a Bigram or Trigram will never see that connection.
4.  **No deep Semantic Meaning:** Even though it captures "not good," it still doesn't understand that "not good" is similar to "bad."
5. **Out of Vocabulary (OOV) :** It still deals with the problem of OOV, as it sees some new words which are not there in the training set, it will simply ignore those words.

### N-gram Range Strategies in CountVectorizer

The `ngram_range` parameter determines how much context your model captures by grouping consecutive words together.

### Comparison of Ranges

| Range | Name | Result | Best For |
| :--- | :--- | :--- | :--- |
| **(1, 1)** | Unigrams | Individual words only. | Simple tasks, very large datasets, or low-memory computers. |
| **(1, 2)** | Unigrams + Bigrams | Individual words and pairs. | **The "Sweet Spot."** Best for most sentiment analysis and classification tasks. |
| **(1, 3)** | Unigrams + Bigrams + Trigrams | Words, pairs, and triplets. | Complex language tasks where 3-word phrases (like "not that bad") are critical. |


### The Trade-off: Context vs. Dimensionality

1.  **Lower Range (1, 1):**
    *   Fewer features (smaller vocabulary).
    *   Faster training.
    *   **Cons:** Misses context. "Not happy" is seen as "not" and "happy."

2.  **Higher Range (1, 3):**
    *   Captures deep context and phrases.
    *   **Cons:** The number of features explodes. If you have 50,000 words, you might end up with millions of features, which can crash your system.


### Recommended "Best Practice" Configuration

When using the IMDB dataset or similar text data, this is the most balanced configuration:

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(
    ngram_range=(1, 2),  # Captures unigrams and bigrams
    max_features=5000    # Limits the vocabulary to the top 5000 most frequent items
)

# This combination gives you the context of Bigrams but prevents 
# the number of features from getting out of control.

## Tf-Idf (Term Frequency - Inverse Document Frequency)

TF-IDF is a statistical measure used to evaluate how important a word is to a document in a collection or corpus. It is an improvement over the Bag of Words (CountVectorizer) because it penalizes words that appear too frequently across all documents (like "the", "is", "and") and gives more weight to unique, meaningful words.

### How it Works

TF-IDF is calculated by multiplying two different metrics:

#### 1. Term Frequency (TF)
This measures how frequently a term occurs in a specific document.
*   **Formula:** (Number of times term T appears in a document) / (Total number of words in that document)
*   **Purpose:** The more a word appears in a document, the more relevant it is to that document.

#### 2. Inverse Document Frequency (IDF)
This measures how important a term is across the entire corpus.
*   **Formula:** log(Total number of documents / Number of documents with term T in it)
*   **Purpose:** If a word appears in almost every document (like "the"), its IDF score will be very low (close to 0). If a word is rare and appears in only a few documents, its IDF score will be high.

**Final Score:** TF-IDF = TF * IDF



### Example Scenario

Imagine you have a collection of 1,000 documents about movies.

*   **Word "the":** Appears in every document. TF might be high, but IDF will be nearly 0. The final TF-IDF score will be very low.
*   **Word "Batman":** Appears many times in 5 specific documents but not in the others. TF will be high for those 5 documents, and IDF will also be high because it is rare in the rest of the 1,000 documents. The final TF-IDF score will be very high.
