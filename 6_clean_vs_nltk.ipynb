```python
import numpy as np
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
```

```python
# nltk
import nltk

# stopwords
from nltk.corpus import stopwords
stop_words = stopwords.words('english')

# stemming
from nltk.stem.porter import PorterStemmer  
st = PorterStemmer()

from cleantext import clean
```

```python
clean('HELLLO !!! this is Ashura, we are happy to welcome you all', punct=True, lowercase=True, stemming=False)
```

```python
clean('Playing, Singing, Dancing', punct=True, lowercase=True)
```

```python
# making a function using both clean and nltk to clean the texts

def change(text):
    text = clean(text, lowercase=True, punct=True, stemming=False, stopwords=False)
    text = nltk.word_tokenize(text)
    

    temp = []
    
    # stopwords
    for word in text:
        if word not in stop_words:
            # stemming
            word = st.stem(word)
            temp.append(word)
    

    # joining temp with a single string
    text = " ".join(temp)
    return text
```

```python
change('hii there my name is king, living here. !!!')
```

### Using Lemmatization instead of stemming to keep the original root word

```python
import nltk

from nltk.tokenize import word_tokenize

from nltk.corpus import stopwords
stop_words = stopwords.words('english')

from nltk.stem import WordNetLemmatizer
lmt = WordNetLemmatizer()

from cleantext import clean
```

```python
help(WordNetLemmatizer)
```

```python
lmt.lemmatize('I am playing cricket right now to make the winning team make proud on me.', pos='v')
```

```python
# printing both the original and root word


sentence = 'I am playing cricket right now to make the winning team make proud on me.'
ans = [(word, lmt.lemmatize(word, pos='v')) for word in sentence.split()]

print(ans)
```

```python
from nltk.stem import WordNetLemmatizer
lmt = WordNetLemmatizer()
```

```python
# making a function that do tokenization, stopword removal, and lemmatization

def transform(text):
    text = clean(text, lowercase=True, punct=True, stopwords=False, stemming=False)
    text = word_tokenize(text)

    temp = []
    for word in text:
        if word not in stop_words:
            # lemmatization
            word = lmt.lemmatize(word, pos='v')
            temp.append(word)
    
    text = temp[:]
    return " ".join(text)
```

```python
transform('Holla !!!, Hope you are fine.')
```

```python
tfidf.fit_transform(['Hello World']).toarray()        
# REMEMBER : while using tfidf you must pass the input in a list format, coz tfidf only takes strings in list format, it doesn't take strings in raw format
# REMEMBER : which using tfidf you must need to convert it into an array, else it will just throw some numbers
```
